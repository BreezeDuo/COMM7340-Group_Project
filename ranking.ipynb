```python
import csv
import networkx as nx
from operator import itemgetter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
with open('Movie_to_Movie_ref.csv', 'r') as file: 
    movie_ref = pd.read_csv(file).iloc[:, 0:2]
with open('Movie_to_Movie_refin.csv', 'r') as file: 
    movie_refin = pd.read_csv(file)
```

```python
ref = movie_ref.groupby('References').size
```

```python
#revice_refpd = pd.DataFrame({'movie':ref.index, 'count':ref.values})
#revice_refpd.to_csv('count_revice.csv',encoding="utf-8")
```

```python
refin= movie_refin.groupby('Referenced in').size
```

```python
#revice_refinpd = pd.DataFrame({'movie':refin.index, 'count':refin.values})
#revice_refinpd.to_csv('count_revice_in.csv',encoding="utf-8")
```

```python
movie_refpd = pd.DataFrame({'movie':ref.index, 'count':ref.values})
movie_refpd=movie_refpd.sort_values(by="count" ) 
movie_refpd.to_csv('movie_refpd_count.csv',encoding="utf-8")
movie_refpd
```

```python
movie_refinpd = pd.DataFrame({'movie':refin.index, 'count':refin.values})
movie_refinpd=movie_refinpd.sort_values(by="count" )
movie_refinpd.to_csv('movie_refinpd_count.csv',encoding="utf-8")
movie_refinpd
```

```python
ref_df= movie_refpd
ref_df['movie'], ref_df['year'] = ref_df['movie'].str.split('(', 1).str

ref_df=ref_df.sort_values(by="year" ) 
ref_df['Movie']=ref_df['movie']+'('+ref_df['year']
```

```python
ref_df= ref_df.groupby('year')
```

```python
year_refpd = pd.DataFrame({'year':ref_df.index, 'count':ref_df.values})
year_refpd=year_refpd.sort_values(by="count" )
year_refpd
year_refpd.to_csv('year_refpd_count_distinct.csv',encoding="utf-8")
```

```python
refin_df= movie_refinpd
refin_df['movie'], refin_df['year'] = refin_df['movie'].str.split('(', 1).str

refin_df=refin_df.sort_values(by="year" ) 
refin_df['Movie']=refin_df['movie']+'('+refin_df['year']
```

```python
refin_df= refin_df.groupby('year').size
```

```python
year_refinpd = pd.DataFrame({'year':refin_df.index, 'count':refin_df.values})
year_refinpd=year_refinpd.sort_values(by="count" )
year_refinpd
year_refinpd.to_csv('year_refinpd_count_distinct.csv',encoding="utf-8")
```
